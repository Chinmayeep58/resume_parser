```python
from pdfminer.high_level import extract_text
import os

def extract_text_from_pdf(pdf_path):
    return extract_text(pdf_path)

sample_text=extract_text_from_pdf("sample_resume.pdf")
print(sample_text[:100])
```

```python
import re

def clean_text(text):
    text=re.sub(r'\n+','\n',text)
    text=re.sub(r' +',' ',text)
    return text.strip()

cleaned=clean_text(sample_text)
print(cleaned)
```

```python
import re

def extract_email(text):
    match = re.search(r'\S+@\S+', text)
    return match.group(0) if match else None

def extract_phone(text):
    match = re.search(r'\+?\d[\d\s\-\(\)]{8,15}', text)
    return match.group(0) if match else None


email = extract_email(sample_text)
phone = extract_phone(sample_text)

print(email, phone)
```

```python
import spacy
nlp=spacy.load('en_core_web_sm')

def extract_name(text):
    doc=nlp(text)
    for ent in doc.ents:
        if ent.label_=="PERSON":
            return ent.text
    return None

name=extract_name(sample_text)
print(name)
```

# matching with jd skills

```python
skill_set=['python','sql','excel','power bi','machine learning', 'data science']

def extract_skills(text,skills=skill_set):
    found=[skill for skill in skills if skill.lower() in text.lower()]
    return list(set(found))

skills=extract_skills(sample_text)
print(skills)
```

# education and degree

```python
keywords=['bachelor','master','b.sc','m.sc','phd','b.e','m.e']

def extract_education(text):
    lines=text.split('\n')
    education=[]
    for line in lines:
        for word in keywords:
            if word.lower() in line.lower():
                education.append(line.strip())
    return education

education=extract_education(sample_text)
print(education)
```

# work experience

```python
def extract_experience(text):
    experience_keywords=['experience','work','internship','employment']
    exp_lines=[]
    lines=text.split('\n')
    for line in lines:
        for word in experience_keywords:
            if word.lower() in line.lower():
                exp_lines.append(line.strip())
    return exp_lines

exp=extract_experience(sample_text)
print(exp)
```

```python
parsed_resume={
    'name':extract_name(cleaned),
    'email':extract_email(cleaned),
    'phone':extract_phone(cleaned),
    'skills':extract_skills(cleaned),
    'education':extract_education(cleaned),
    'experience':extract_experience(cleaned)
}

print(parsed_resume)
```
